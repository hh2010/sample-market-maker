```python
import pandas as pd
import json
import os
```

```python
df = pd.DataFrame()

with open('./xbt_usd.txt') as fn:
    for line in fn.readlines():
        if line[0] == '{' and 'success' not in line and 'info' not in line:
#             print(line)
            line = json.loads(line.strip())
            
            table = line['table']
            action = line['action']
            data = line['data']
            
            df_temp = pd.DataFrame([table, action, data]).transpose()
            
            df = pd.concat([df, df_temp], ignore_index=True)
```

```python
dfs = []
for df_tmp in df[2].values:
    dfs.append(pd.DataFrame(df_tmp))
```

```python
df_data = pd.concat(dfs, ignore_index=True)
```

```python
df_final= pd.concat([df, df_data], axis=1).drop(2, axis=1)
```
